```python
import numpy as np
import matplotlib.pyplot as plt
from collections import namedtuple
```

```python
ax = plt.subplot()
consumers = range(1, 11)
wtp = (98, 72, 41, 38, 29, 21, 17, 12, 11, 10)
price = 25
ax.bar(consumers, wtp, label = "consumer surplus", color = "darkblue", alpha = 0.8)
ax.plot((0, 12), (price, price), lw=2, label="price $p$")
ax.bar(consumers, [min(w, price) for w in wtp], color="black", alpha=0.6)
ax.set_xlim(0,12)
ax.set_xticks(consumers)
ax.set_ylabel("willingless to pay, price")
ax.set_xlabel("consumer, quantity")
ax.legend()
plt.show()
```
